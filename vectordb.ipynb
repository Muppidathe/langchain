```python
from langchain.document_loaders import PyMuPDFLoader
loader=PyMuPDFLoader('Resume_v1.pdf')
doc=loader.load()
doc
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
docs=splitter.split_documents(doc)
docs
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings=OllamaEmbeddings(model='mxbai-embed-large')
embedded_docs=embeddings.embed_documents(docs)
embedded_docs
```

### faiss DB

```python
from langchain_community.vectorstores import FAISS
db=FAISS.from_documents(docs,embedding=embeddings)
db
```

```python
db.similarity_search('strong foundation in programming and a deep passion')
```

```python
db.similarity_search_with_score('strong foundation in programming and a deep passion')
```

## retriver

```python
retriver=db.as_retriever(search_kwargs={'k': 6})
retriver.invoke('strong foundation in programming and a deep passion')
```

### Chroma db

```python
from langchain_community.vectorstores import Chroma
#from langchain_chroma import Chroma
db=Chroma.from_documents(docs,embeddings)
db
```

```python
db.similarity_search('BCA')
```
